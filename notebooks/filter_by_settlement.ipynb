# Filter by Settlement (特定の農業集落のデータだけにする)

```python
import geopandas as gpd

# Load the GeoJSON file
file_path = '../src/models/area_optimization/data/processed/merged_geojson/merged_polygon_and_farm_pin_and_settlement.geojson'
gdf = gpd.read_file(file_path)

gdf.head()
```

```python
# Settlementの種類を確認
settlement_types = gdf['Settlement_name'].unique()
print(settlement_types)
```

```python
Settlement_name = '筑地'
```

```python
# 指定されたデータを抽出
filtered_gdf = gdf[gdf['Settlement_name'] == Settlement_name]
print(len(filtered_gdf))
filtered_gdf.head()
```

```python
# データを保存
output_file_path = f'../notebooks/data/geojson_filtered_by_settlement/{Settlement_name}.geojson'
filtered_gdf.to_file(output_file_path, driver='GeoJSON')
```
